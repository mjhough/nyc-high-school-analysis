# Read in the data

```python
import pandas as pd
import numpy
import re

data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]

data = {}

for f in data_files:
    d = pd.read_csv("schools/{0}".format(f))
    data[f.replace(".csv", "")] = d
```

# Read in the surveys

```python
all_survey = pd.read_csv("schools/survey_all.txt", delimiter="\t", encoding='windows-1252')
d75_survey = pd.read_csv("schools/survey_d75.txt", delimiter="\t", encoding='windows-1252')
survey = pd.concat([all_survey, d75_survey], axis=0)

survey["DBN"] = survey["dbn"]

survey_fields = [
    "DBN", 
    "rr_s", 
    "rr_t", 
    "rr_p", 
    "N_s", 
    "N_t", 
    "N_p", 
    "saf_p_11", 
    "com_p_11", 
    "eng_p_11", 
    "aca_p_11", 
    "saf_t_11", 
    "com_t_11", 
    "eng_t_11", 
    "aca_t_11", 
    "saf_s_11", 
    "com_s_11", 
    "eng_s_11", 
    "aca_s_11", 
    "saf_tot_11", 
    "com_tot_11", 
    "eng_tot_11", 
    "aca_tot_11",
]
survey = survey.loc[:,survey_fields]
data["survey"] = survey
```

# Add DBN columns

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

def pad_csd(num):
    string_representation = str(num)
    if len(string_representation) > 1:
        return string_representation
    else:
        return "0" + string_representation
    
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]
```

# Convert columns to numeric

```python
cols = ['SAT Math Avg. Score', 'SAT Critical Reading Avg. Score', 'SAT Writing Avg. Score']
for c in cols:
    data["sat_results"][c] = pd.to_numeric(data["sat_results"][c], errors="coerce")

data['sat_results']['sat_score'] = data['sat_results'][cols[0]] + data['sat_results'][cols[1]] + data['sat_results'][cols[2]]

def find_lat(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lat = coords[0].split(",")[0].replace("(", "")
    return lat

def find_lon(loc):
    coords = re.findall("\(.+, .+\)", loc)
    lon = coords[0].split(",")[1].replace(")", "").strip()
    return lon

data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(find_lat)
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(find_lon)

data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")
```

# Condense datasets

```python
class_size = data["class_size"]
class_size = class_size[class_size["GRADE "] == "09-12"]
class_size = class_size[class_size["PROGRAM TYPE"] == "GEN ED"]

class_size = class_size.groupby("DBN").agg(numpy.mean)
class_size.reset_index(inplace=True)
data["class_size"] = class_size

data["demographics"] = data["demographics"][data["demographics"]["schoolyear"] == 20112012]

data["graduation"] = data["graduation"][data["graduation"]["Cohort"] == "2006"]
data["graduation"] = data["graduation"][data["graduation"]["Demographic"] == "Total Cohort"]
```

# Convert AP scores to numeric

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data["ap_2010"][col] = pd.to_numeric(data["ap_2010"][col], errors="coerce")
```

# Combine the datasets

```python
combined = data["sat_results"]

combined = combined.merge(data["ap_2010"], on="DBN", how="left")
combined = combined.merge(data["graduation"], on="DBN", how="left")

to_merge = ["class_size", "demographics", "survey", "hs_directory"]

for m in to_merge:
    combined = combined.merge(data[m], on="DBN", how="inner")

combined = combined.fillna(combined.mean())
combined = combined.fillna(0)
```

# Add a school district column for mapping

```python
def get_first_two_chars(dbn):
    return dbn[0:2]

combined["school_dist"] = combined["DBN"].apply(get_first_two_chars)
```

# Find correlations

```python
correlations = combined.corr()
correlations = correlations["sat_score"]
print(correlations)
```

# Plotting race correlations

```python
combined.corr()['sat_score'][['white_per', 'asian_per', 'black_per', 'hispanic_per']].plot.bar()
```

The above bar plot shows that both the percentage of white and asian students is positively correlated to SAT scores. On the other hand, there is a negative correlation between SAT scores and the percentage of black and hispanic students. 

```python
combined.plot.scatter('hispanic_per', 'sat_score')
plt.xlim(-10,110)
```

The above scatter plot, proves the finding in the bar plot. This shows that having a lower percentage of hispanic students at a school often results in higher SAT scores.

However, it is evident that the SAT scores are very much clustered around the same range throughout, apart from the outliers at the 0% and 100% hispanic regions. 

```python
high_hispanic_schools = combined[combined['hispanic_per'] > 95]
high_hispanic_schools
```

It appears that the majority of these schools are either international/multicultural schools, or schools geared to specific skills. These students often speak another language other than English, which would explain the low reading/writing scores in comparison to Math, as well as the low SAT scores.

```python
low_hispanic_schools = combined[combined['hispanic_per'] < 10]
low_hispanic_schools
```

Most of these schools are not focused on international students. This shows in the SAT scores, where reading, writing and math all score relatively the same on average.

# Gender and SAT scores

```python
combined.corr()['sat_score'][['male_per', 'female_per']].plot.bar()
```

There are slight correlations between the percentage of males and females, and SAT scores. A higher percentage of male students results in a negative correlation, while a higher percentage of female students results in a positive correlation.

These correlations are not significant however.

```python
combined.plot.scatter('female_per', 'sat_score')
plt.xlim(-10,110)
```

The scatter plot shows little difference in SAT score compared with the percentage of females at a school. Schools with 100% females did perform better than schools with 0% females, contributing to the slight correlation.

It is evident that the best SAT scores come from schools where there are an even number of males and females. This is where the majority of schools sit, however.

```python
high_female_schools = combined[(combined['female_per'] > 60) & (combined['sat_score'] > 1700)]
high_female_schools
```

The majority of the schools with a high percentage of females appear to be schools with very high acceptance standards. This would explain higher SAT scores.
